Q1. What is the syntax for adding an element to a list in Python?

Answer:


In Python, there are a few different ways to add an element to a list. Here are the commonly used methods:

1. Using the `append()` method:
The `append()` method is used to add an element to the end of a list. It modifies the list in-place, meaning it directly affects the original list.

Syntax:
```python
list_name.append(element)
```
Example:
```python
my_list = [1, 2, 3]
my_list.append(4)
print(my_list)
```
Output:
```
[1, 2, 3, 4]
```

2. Using the `insert()` method:
The `insert()` method is used to insert an element at a specific position in the list. It shifts the existing elements to the right.

Syntax:
```python
list_name.insert(index, element)
```
Example:
```python
my_list = [1, 2, 3]
my_list.insert(1, 4)
print(my_list)
```
Output:
```
[1, 4, 2, 3]
```

3. Using the concatenation operator (`+`):
You can use the concatenation operator (`+`) to combine two lists, effectively adding elements from one list to another.

Syntax:
```python
new_list = list1 + list2
```
Example:
```python
my_list = [1, 2, 3]
new_element = [4]
new_list = my_list + new_element
print(new_list)
```
Output:
```
[1, 2, 3, 4]
```

Q2. What is the difference between remove() and pop() functions in Python?

Answer:

In Python, both the `remove()` and `pop()` functions are used to remove elements from a list, but they have different behaviors:

1. `remove()` function:
The `remove()` function is used to remove the first occurrence of a specified value from a list.

Syntax:
```python
list_name.remove(value)
```
- The `value` is the element to be removed from the list.

Example:
```python
my_list = [1, 2, 3, 2]
my_list.remove(2)
print(my_list)
```
Output:
```
[1, 3, 2]
```

2. `pop()` function:
The `pop()` function is used to remove an element from a list based on its index. It returns the removed element.

Syntax:
```python
list_name.pop(index)
```

Example:
```python
my_list = [1, 2, 3, 4]
removed_element = my_list.pop(1)
print(removed_element)
print(my_list)
```
Output:
```
2
[1, 3, 4]
```

The key difference between `remove()` and `pop()` is that `remove()` removes the first occurrence of a specified value, whereas `pop()` removes an element based on its index and returns it. Additionally, `pop()` can be used to remove and retrieve the last element of a list if no index is specified.

Q3. Write a Python code to sort a list in descending order.

Answer:

To sort a list in descending order, we can use the `sort()` method with the `reverse` parameter set to `True`.     

Syntax:
```python
list_name.sort(reverse=True)
```

Example:
```python
my_list = [1, 3, 2]
my_list.sort(reverse=True)
print(my_list)
```

Output:
```
[3, 2, 1]
```

Q4. Write a Python code to count the number of occurrences of an element in a list.

Answer:

To count the number of occurrences of an element in a list, we can use the `count()` method.

Syntax:
```python  
list_name.count(element)
```

Example:
```python  
my_list = [1, 2, 3, 2]
count = my_list.count(2)
print(count)
```

Output:
```
2
```

Q5. Write a Python code to reverse a list.

Answer:

To reverse a list, we can use the `reverse()` method.

Syntax:
```python   
list_name.reverse()
```

Example:
```python
my_list = [1, 2, 3]
my_list.reverse()
print(my_list)
```

Output:
```
[3, 2, 1]
```
